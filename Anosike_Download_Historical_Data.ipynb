```python
import yfinance as yf
import math
import pandas_datareader as web
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
import datetime
plt.style.use("fivethirtyeight")

import requests
import random
import time
import json
from bs4 import BeautifulSoup
from google.colab import drive
import os
```

```python
amazon = yf.Ticker("AMZN")
#Set the time range
amazon_hist = amazon.history(start=datetime.datetime(2010, 1, 1),end=datetime.datetime.today())
```

```python
amazon_hist.head(20)
```

```python


# Mount Google Drive
drive.mount('/content/drive')
drive_path = '/content/drive/My Drive/TimeSeries/'

# Ensure amazon_hist exists and has data
if 'amazon_hist' in locals() and not amazon_hist.empty:
    df = amazon_hist.reset_index()  # Move Date index to a column

    # Convert Date column to timezone-na√Øve format
    df['Date'] = df['Date'].dt.tz_localize(None)

    # Save DataFrame to CSV and Excel
    df.to_csv(os.path.join(drive_path, 'historical_data.csv'), index=False)
    df.to_excel(os.path.join(drive_path, 'historical_data.xlsx'), index=False)

    print("Data has been saved successfully to Google Drive.")
else:
    print("No data to save.")

```
